## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] > 0]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions > 0)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# Your discovery here
### My Week 1 Discovery: The Volume-Engagement Trade-off

# **The Observation:**
# I analyzed the relationship between `content_type` and performance metrics. While `keyword articles` generate significantly higher median `impressions_90d`, they do not demonstrate the highest `engagement_rate`.

# **The Insight:**
# There is a clear decoupling between reach (volume) and quality (engagement). Relying solely on `keyword articles` for growth may be a "leaky bucket" strategy—the content drives traffic but fails to retain the user.

# **Conclusion:**
# Performance optimization should not be based on search volume alone. To maximize value, we need to balance content types that drive discovery with those that actually sustain user engagement.
```

```python
print(df.columns)
insight = df.groupby("content_type")["impressions_90d"].median().sort_values(ascending=False)
print(insight)
```

```python
# Investigating if content_type influences quality (engagement_rate)
# Let's see if the type of content that gets the most traffic (keyword articles)
# is also the type that keeps people engaged.

engagement_by_type = df.groupby("content_type")["engagement_rate"].median().sort_values(ascending=False)

print("Median engagement_rate by content_type:")
print(engagement_by_type.round(4).to_string())

# Observation/Reflection:
# Compare this to your previous finding.
# Do 'keyword articles' have the highest engagement, or do users prefer a different type?
```

My Week 1 Discovery:
I observed that while keyword articles drive the highest median traffic (impressions), they do not show the highest engagement rates compared to comparison articles. This suggests that volume and user intent (as captured by engagement_rate) are decoupled by content_type.
