## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Show traffic volume difference
median_impressions = df.groupby("content_type")["impressions_90d"].median()
print(f"Median Impressions:\n{median_impressions}")

# Show decline rate
decline_rate = df["trend_direction"].value_counts(normalize=True)["down"]
print(f"\nOverall percentage of declining pages: {decline_rate:.2%}")
```
